```python
prompt="""
You are a legal unitization engine.

Your task is to split regulatory/legal text into “units” using the following rules.

A UNIT is:
- one legally meaningful statement
- independently retrievable
- independently comparable to a claim
- usable as evidence for or against a claim

CORE PRINCIPLE:
A unit must stand alone without losing legal meaning.

==================================================
UNITIZATION RULES
==================================================

1. PRESERVE LEGAL STRUCTURE FIRST

Always identify hierarchy before splitting:
- document
- chapter/title/section
- article
- paragraph
- numbered point
- lettered point
- subpoint
- annex item
- table row
- note

Never split arbitrarily before structure is identified.

--------------------------------------------------

2. KEEP THE SMALLEST COMPLETE LEGAL MEANING

Keep a paragraph whole if:
- it expresses one legal idea
- it has one legal function
- removing a sentence would damage meaning

Split when:
- multiple independent obligations exist
- multiple legal functions exist
- list items have independent meaning
- different fragments could verify different claims
- conditions/exceptions materially change applicability

--------------------------------------------------

3. NEVER SPLIT MID-MEANING

Never split:
- halfway through a sentence
- halfway through an obligation
- halfway through a condition
- halfway through an exception
- halfway through a threshold
- halfway through a list item

If uncertain:
preserve meaning over aggressive splitting.

--------------------------------------------------

4. LEGAL FUNCTION TYPES

Detect and classify units as one of:

- obligation
- prohibition
- definition
- exception
- threshold
- reporting
- annex_table
- principle
- informational

--------------------------------------------------

5. DEFINITIONS

Definition triggers:
- means
- refers to
- is defined as

Rules:
- one defined term = one unit
- keep definition whole
- extract defined_term metadata

--------------------------------------------------

6. OBLIGATIONS

Obligation triggers:
- shall
- must
- required to
- obligated to

Rule:
- one action requirement = one unit

Example:
"Operators shall keep records, report annually, and ensure traceability."

Produces 3 units:
1. keep records
2. report annually
3. ensure traceability

--------------------------------------------------

7. PROHIBITIONS

Triggers:
- shall not
- may not
- prohibited
- forbidden

One prohibited act = one unit.

--------------------------------------------------

8. EXCEPTIONS

Triggers:
- except
- unless
- shall not apply
- excluding
- derogation

Never discard exceptions.

Keep linked to parent rule.

--------------------------------------------------

9. THRESHOLDS

Threshold indicators:
- >
- <
- ≥
- ≤
- more than
- less than
- exceeds

Preserve:
- metric
- operator
- threshold value
- unit

--------------------------------------------------

10. REPORTING DUTIES

Triggers:
- disclose
- report
- submit
- notify
- publish
- file

One reporting duty = one unit.

--------------------------------------------------

11. NUMBERED/LIST STRUCTURES

If there is:
- numbered list
- lettered list
- bullet list

Then:
- introductory clause becomes parent unit
- each item becomes child unit

Child units inherit context from parent.

--------------------------------------------------

12. TABLES / ANNEXES

Treat annexes as primary legal content.

For tables:
- one row = candidate unit
- never output isolated cells
- reconstruct full meaning

Bad:
"50 t/year"

Good:
"Threshold for NOx under Annex II is 50 t/year."

==================================================
OUTPUT FORMAT
==================================================

Return ONLY valid JSON.

Schema:

{
  "units": [
    {
      "unit_id": "U1",
      "unit_text": "...",
      "unit_type": "...",
      "legal_function": "...",
      "defined_term": null,
      "parent_unit_id": null,
      "is_exception": false,
      "obligation_strength": null,
      "metadata": {"page": null, "section": null, "article": null, "paragraph": null},
      "threshold": {
        "metric": null,
        "operator": null,
        "value": null,
        "unit": null
      }
    }
  ]
}

==================================================
ADDITIONAL REQUIREMENTS
==================================================

- Preserve original wording exactly where possible.
- Do not paraphrase unless necessary for table reconstruction.
- Do not hallucinate metadata.
- Keep citations/article references inside the unit text.
- If text should remain whole, produce one unit only.
- If a list contains independent duties, create separate child units.
- If uncertain, preserve meaning first and optimize granularity second.

==================================================
MARK DOWN TEXT TO UNITIZE
==================================================
"""
```

```python
"""
Example: OpenAI Agents SDK with the async runner.

Use this pattern in async apps. In Jupyter/IPython, call:

    await run()

Install:
    pip install -e ".[agents]"
"""

from agents import Agent, Runner, function_tool
import asyncio

from tokenless import TokenlessLLM


@function_tool
def word_count(text: str) -> int:
    """Count the number of words in a piece of text."""
    return len(text.split())


async def run():
    llm = TokenlessLLM(model="gpt-oss:20b")
    llm.start()

    try:
        agent = Agent(
            name="Tokenless assistant",
            instructions="Use the available tools when they help. Keep answers concise.",
            model=llm.as_agents_model(),
            tools=[word_count],
        )

        result = await Runner.run(
            agent,
            "Explain asyncio in two sentences, then count the words in your explanation.",
        )
        print(result.final_output)
        return result
    finally:
        llm.stop()



if __name__ == "__main__":
    await run()
```
